### StatsModels 라이브러리를 활용한 선형회귀 모델링과 진단.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.datasets import load_boston
%matplotlib inline
```

#### 1. 'Boston' dataset을 Scikit-Learn에서 가져와서 DataFrame으로 변환해 둔다.

```python
data = load_boston()
```

```python
# 설명변수.
X = data['data']
header = data['feature_names']
```

```python
# 반응변수.
Y = data['target']
Y = Y.reshape(-1, 1)
```

```python
# DataFrame으로 변환.
df = pd.DataFrame(np.append(X,Y,axis = 1))
df.columns = list(header) + ['PRICE']
```

```python
df.head(5)
```

#### 2. R 스타일의 수식을 사용하여 모델링 한다.

```python
# 모든 설명변수를 사용한 모형.
myModel1 = smf.ols(formula = "PRICE ~ CRIM + ZN + INDUS + CHAS + NOX + RM + AGE + DIS + RAD + TAX + PTRATIO + B + LSTAT ", data=df)
result1 = myModel1.fit()
result1.summary()
```

**NOTE: INDUS와 AGE는 유의한 설명 변수가 아니다. Adjusted $R^2 = 0.734$ and $AIC = 3026$.** 

```python
# 'INDUS'와 'AGE'를 제거한 모형.
myModel2 = smf.ols(formula = "PRICE ~ CRIM + ZN + CHAS + NOX + RM + DIS + RAD + TAX + PTRATIO + B + LSTAT ", data=df)
result2 = myModel2.fit()
result2.summary()
```

**NOTE: 모든 설명 변수는 유의하다. Adjusted $R^2 = 0.735$ and $AIC = 3022$.** 

**CONCLUSION: 두 번째 모델이 첫 번째 모델보다 더 최적화 된 듯 하다.**

#### 3. VIF를 계산해 본다.

```python
myDict = {"변수":[], "VIF":[]}
for i, feature in enumerate(myModel2.exog_names):
    if feature != "Intercept":
        myDict["변수"] += [feature]
        myDict["VIF"] += [variance_inflation_factor(myModel2.exog, i)]
pd.DataFrame(myDict)
```
